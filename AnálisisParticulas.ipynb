```python
plt.figure(figsize=(10, 6))
plt.scatter(df['eta1'], df['eta2'], alpha=0.5)
plt.xlabel('Pseudorapidez (η1)')
plt.ylabel('Pseudorapidez (η2)')
plt.title('Correlación entre η1 y η2')
plt.grid()
plt.show()
```

```python
high_pt_events = df[(df['pt1'] > 20) | (df['pt2'] > 20)]
print("Eventos con pT > 20 GeV:\n", high_pt_events[['Run', 'Event', 'pt1', 'pt2']])
```

```python
# Diferencia angular entre partículas
delta_phi = np.abs(df['phi1'] - df['phi2'])
delta_phi = np.where(delta_phi > np.pi, 2*np.pi - delta_phi, delta_phi)  # Ajuste a [0, π]

plt.figure(figsize=(10, 6))
plt.hist(delta_phi, bins=50)
plt.xlabel('Δφ [radianes]')
plt.ylabel('Frecuencia')
plt.title('Diferencia Azimutal entre Partículas')
plt.grid()
plt.show()
```

```python
# Asumimos partículas con masa despreciable (e.g., muones)
df['invariant_mass'] = np.sqrt(2 * df['pt1'] * df['pt2'] * (np.cosh(df['eta1'] - df['eta2']) - np.cos(df['phi1'] - df['phi2'])))

plt.figure(figsize=(10, 6))
plt.hist(df['invariant_mass'], bins=100, range=(0, 200))
plt.xlabel('Masa Invariante [GeV]')
plt.ylabel('Frecuencia')
plt.title('Distribución de Masa Invariante')
plt.grid()
plt.show()
```

```python
fig = go.Figure(data=go.Scatter(
    x=df['eta1'],
    y=df['phi1'],
    mode='markers',
    marker=dict(
        size=5,
        color=df['pt1'],
        colorscale='Viridis',
        showscale=True
    )
))
fig.update_layout(
    title='Mapa de η vs φ (Partícula 1)',
    xaxis_title='Pseudorapidez (η)',
    yaxis_title='Ángulo Azimutal (φ)'
)
fig.show()
```
